```python
"""
Let say you are a teacher and you have different student records containing id of a student and the marks list 
in each subject where different students have taken different number of subjects. All these records are in hard copy.
You want to enter all the data in computer and want to compute the average marks of each student and display
"""

def getDateFromUser():
    D = {}
    while True:
        studentId = input('Enter Student ID: ')
        marksList = input('Enter the marks by comma seperated values: ')
        moreStudents = input('Enter "no" to quit insertion: ')
        if studentId in D:
            print(studentId," is already exist")
        else:
            D[studentId] = marksList.split(",")
        if moreStudents.lower() == "no":
            return D
        
        
        
def getAvgMarks(D):
    avgMarks = {}
    for x in D:
        L = D[x]
        s = 0
        for marks in L:
            s += int(marks)
        avgMarks[x] = s/len(L)
    return avgMarks
```

```python
studentData = getDateFromUser()
```

```python
studentData
```

```python
avgMarks = getAvgMarks(studentData)
```

```python
for x in avgMarks:
    print("Student",x,"got avg Marks as:",avgMarks[x])
```
